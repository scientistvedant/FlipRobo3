### 7 Top 10 Gaming Laptop

```python
#creating the driver
driver = webdriver.Chrome('chromedriver2.exe')

#setting url
url = 'https://www.digit.in/'

#attaching the url to driver
driver.get(url)

#search box
search = driver.find_element_by_xpath("//div[@class='search']/a").click()

#input
input = driver.find_element_by_xpath("//input[@type='search']")

# sending keys
input.send_keys('Top 10 gaming laptop')

#clicking
input.send_keys(Keys.ENTER)

#creating the URL lists
urls = []

for u in driver.find_elements_by_xpath("//div[@class='searchPage']/a"):
    urls.append(u.get_attribute('href'))
    
#clicking on the first option
click_top1 = driver.find_element_by_xpath("//div[@class='searchProduct-desc']").click()


# empty dataset
Price = []
Brand = []

for i in urls:
    driver.get(i)
    
    for j in driver.find_elements_by_xpath("//div[@class='Block-price']"):
        Price.append(j.text)
        
    for k in driver.find_elements_by_xpath("//div[@class='heading-wraper']"):
        Brand.append(k.text)

# create the dataframe
Top_10 = pd.DataFrame({})
Top_10['Brand'] = Brand
Top_10['Price'] = Price   
Top_10
```

### 6 Funding Deals for second Quarter (trak.in)

```python
#creating the driver
driver = webdriver.Chrome('chromedriver2.exe')

#approaching to the website
driver.get('https://trak.in/')
```

```python
#find the funding details using id
select_fund = driver.find_element_by_xpath('//*[@id="menu-item-51510"]')

#click on the object
select_fund.click()
```

```python
months = []
Start_up = []
Industries = []
Sub_verticals = []
Locations = []
Investors = []
Investors_type = []
Amounts = []

#fetching the details via xpath
month = driver.find_elements_by_xpath("//td[@class='column-2']")
SP = driver.find_elements_by_xpath("//td[@class='column-3']")
Industry = driver.find_elements_by_xpath("//td[@class='column-4']")
Sub_vertical = driver.find_elements_by_xpath("//td[@class='column-5']")
Location = driver.find_elements_by_xpath("//td[@class='column-6']")
Investor = driver.find_elements_by_xpath("//td[@class='column-7']")
Investor_type = driver.find_elements_by_xpath("//td[@class='column-8']")
Amount = driver.find_elements_by_xpath("//td[@class='column-9']")

#plaing the fetched details into the empty dataset
for i in month:
    months.append(i.text)
    
for i in SP:
    Start_up.append(i.text)
    
for i in Industry:
    Industries.append(i.text)
    
for i in Sub_vertical:
    Sub_verticals.append(i.text)
    
for i in Location:
    Locations.append(i.text)
    
for i in Investor:
    Investors.append(i.text)
    
for i in Investor_type:
    Investors_type.append(i.text)
    
for i in Amount:
    Amounts.append(i.text)
```

```python
# crating the dataframe
Financial_Report= pd.DataFrame({})
Financial_Report['Month'] = months
Financial_Report['Start_up'] = Start_up
Financial_Report['Industry'] = Industries
Financial_Report['Sub_verticals0'] = Sub_verticals
Financial_Report['Locations'] = Locations
Financial_Report['Investors'] = Investors
Financial_Report['Investors_type'] = Investors_type
Financial_Report['Amounts'] = Amounts
```

```python
# correcting the date 05/072018 format to normal dateformat
Financial_Report['Month'] = Financial_Report['Month'].replace('05/072018','05/07/2018')
```

```python
# seperating the details monthwise by creating unique dataframes
Financial_Report['Month'] = pd.to_datetime(Financial_Report['Month'])
Financial_Report7 = Financial_Report[Financial_Report['Month'].dt.month == 7] 
Financial_Report8 = Financial_Report[Financial_Report['Month'].dt.month == 8]
Financial_Report9 = Financial_Report[Financial_Report['Month'].dt.month == 9]
```

```python
# concatenating all the datasets to create report for second quarter
SecondQuarter = pd.concat([Financial_Report7, Financial_Report8, Financial_Report9], axis=0)
SecondQuarter
```

### 4. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps

```python
# creating the driver
driver = webdriver.Chrome('chromedriver2.exe')

#reaching the website via url
driver.get("https://www.flipkart.com/")
time.sleep(2)

# search box
box = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
time.sleep(2)

#sending input
box.send_keys('pixel 4A')
box.send_keys(Keys.ENTER)
```

```python
Brand = []
#SmartPhone_name = []
#Color = []
RAM_ROM = []
Camera_Resolution = []
Display_Size = []
#Display_resolution = []
Processor = []
Processor_Cores = []
Battery_Capacity = []
Price = []
Product_URL = []

Brands = driver.find_elements_by_xpath('//div[@class="_4rR01T"]')
ROM = driver.find_elements_by_xpath('//li[@class="rgWa7D"][1]')
Display = driver.find_elements_by_xpath('//li[@class="rgWa7D"][2]')
Camera = driver.find_elements_by_xpath('//li[@class="rgWa7D"][3]')
Battery = driver.find_elements_by_xpath('//li[@class="rgWa7D"][4]')
Processors = driver.find_elements_by_xpath('//li[@class="rgWa7D"][5]')
price = driver.find_elements_by_xpath('//div[@class="_30jeq3 _1_WHN1"]')
for i in driver.find_elements_by_xpath("//div[@class='_2kHMtA']/a"):
    Product_URL.append(i.get_attribute('href'))


for i in Brands:
    Brand.append(i.text)
    
for i in ROM:
    RAM_ROM.append(i.text)

for i in Camera:
    Camera_Resolution.append(i.text)

for i in Display:
    Display_Size.append(i.text)

for i in Battery:
    Battery_Capacity.append(i.text)

for i in Processors:
    Processor.append(i.text)
    
for i in price:
    Price.append(i.text)
```

```python
#splitting the Brand name
Brand_Name = []
for i in range(len(Brand)):
    Brand_Name.append(Brand[i].split(sep = '(')[0])

#splitting the Smartphone name
SmartPhone_name = []
for i in range(len(Brand_Name)):
    SmartPhone_name.append(Brand_Name[i].split(sep = ' ')[1:])
    
Brand_Names = []
for i in range(len(Brand)):
    Brand_Names.append(Brand[i].split(sep = ' ')[0])
    
#splitting the color
Color = []
for i in range(len(Brand)):
    Color.append(Brand[i].split(sep = ' ')[3:7])
    
#splitting the RAM from RAM_ROM
RAM = []
for i in range(len(RAM_ROM)):
    RAM.append(RAM_ROM[i].split(sep = '|')[0])

#splitting the ROM from RAM_ROM
ROM = []
for i in range(len(RAM_ROM)):
    ROM.append(RAM_ROM[i].split(sep = '|')[1])
    
#splitting the Primary Camera 
Primary_Camera = []
for i in range(len(Camera_Resolution)):
    Primary_Camera.append(Camera_Resolution[i].split(sep = '|')[0])
    
#splitting the Secondary Camera 
Secondary_Camera = []
for i in range(len(Camera_Resolution)):
    Secondary_Camera.append(Camera_Resolution[i].split(sep = '|')[-1])

#splitting the Display Size
Display_sizes = []
for i in range(len(Display_Size)):
    Display_sizes.append(Display_Size[i].split(sep = ')')[0])
    
#splitting the Display Resolution
Display_Resolution = []
for i in range(len(Display_Size)):
    Display_Resolution.append(Display_Size[i].split(sep = ')')[1])
    
#splitting the Processor
Processors = []
for i in range(len(Processor)):
    Processors.append(Processor[i].split(sep = ' ')[0:2])
    
#splitting the Processor_cores
Processors_cores = []
for i in range(len(Processor)):
    Processors_cores.append(Processor[i].split(sep = ' ')[-2:])
    

#creating the dataset    
Mobile_Brand = pd.DataFrame({})
Mobile_Brand['Brand_Name'] = Brand_Names
Mobile_Brand['Smartphone_name'] = SmartPhone_name
Mobile_Brand['Color'] = Color
Mobile_Brand['RAM'] = RAM
Mobile_Brand['ROM'] = ROM
Mobile_Brand['Primary_Camera'] = Primary_Camera
Mobile_Brand['Secondary_Camera'] = Secondary_Camera
Mobile_Brand['Display_Size'] = Display_sizes
Mobile_Brand['Display_Resolution'] = Display_Resolution
Mobile_Brand['Processor'] = Processors
Mobile_Brand['Processor_cores'] = Processors_cores
Mobile_Brand['Battery_Capacity'] = Battery_Capacity
Mobile_Brand['Price'] = Price
Mobile_Brand['URL'] = Product_URL
```
